# Pie Chart

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c", "c++", "Java", "Python"]

ex=[0.3, 0.0, 0.0, 0.0]
c=['r', 'b', 'g', 'y']

plt.pie(x, labels=y, explode=ex, colors=c, autopct="%0.1f%%")   #labels= for 'y' labels / explode= for move to outside

plt.show()   #show()= To show the piechart
```

```python
x=[10,20,30,40]
y=["c", "c++", "Java", "Python"]

ex=[0.3, 0.0, 0.0, 0.0]
c=['r', 'b', 'g', 'y']

plt.pie(x, autopct="%0.1f%%", shadow=True)   #autopact= for percercentages

plt.show()
```

```python
x=[10,20,30,40]
y=["c", "c++", "Java", "Python"]

ex=[0.3, 0.0, 0.0, 0.0]
c=['r', 'b', 'g', 'y']

plt.pie(x, labels=y, colors=c, radius=1.5, autopct="%0.1f%%")   #autopact= for percercentages

plt.show()
```

```python
x=[10,20,30,40]
y=["c", "c++", "Java", "Python"]

ex=[0.3, 0.0, 0.0, 0.0]


plt.pie(x, labels=y, labeldistance=1.5, startangle=90
         , textprops={"fontsize":15}, counterclock=False) #labeldistance= for moving labels near or far 
                                                       #startangle= for decide from where to start the piechart

plt.show()   
```

```python
x=[10,20,30,40]
y=["c", "c++", "Java", "Python"]

ex=[0.3, 0.0, 0.0, 0.0]


plt.pie(x, labels=y, autopct="%0.1f%%",shadow=True, wedgeprops={"linewidth":2, "edgecolor":"m"},
          center=(1,3), rotatelabels=True) 

plt.title("Pie Chart.ipynb", fontsize=20)
plt.legend(loc=2)  #loc= for location of legend, i.e 2 for Java

plt.show()   
```
