# Text in Graph

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,1,4,2,6]

plt.plot(x, y)
plt.title('%matplotlib')

plt.text(2,5, "python", fontsize=15, style="italic", bbox={"facecolor":"red"})

plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,4,2,6]

plt.plot(x, y)
plt.title('%matplotlib')

plt.text(2,5, "python", fontsize=15, style="italic", bbox={"facecolor":"red"})

plt.annotate("python", xy=(3,2), xytext=(4,4), arrowprops=dict(facecolor="black", shrink=100))

plt.legend(["up"], loc=2, facecolor="red", edgecolor="g")

plt.show()
```
