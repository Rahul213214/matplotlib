```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y)
plt.fill_between(x, y)
plt.savefig("plot", dpi=1000)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y)
plt.fill_between(x, y)
plt.savefig("plot1", dpi=500, facecolor="g")

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y)
plt.fill_between(x, y)
plt.savefig("plot3.pdf", dpi=500, facecolor="g") #.pdf / .jpg

plt.show()
```
