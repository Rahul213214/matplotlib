# Stack Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area=[3,5,9,4,8]
area1=[2,3,4,5,6]
area2=[4,1,5,9,4]

l=["area", "area1", "area2"]

plt.stackplot(x, area, area1, area2, labels=l, colors=["r", "g","m"])

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.grid()
plt.legend(loc=2)

plt.show()
```
