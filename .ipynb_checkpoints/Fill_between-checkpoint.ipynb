# Fill_Between

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y)
plt.fill_between(x, y)

plt.title("Matplotlib")

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y)
plt.fill_between(x=[2, 4], y1=2, y2=4, color="g")

plt.title("Matplotlib")

plt.show()
```
