# Scatter Plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
num=[2,3,4,2,4,5,6]

colors=["r","b","g","y","black","brown","orange"]
sizes=[180,190,2800,155,380,2200,130]
plt.scatter(day, num, color=colors, s=sizes, alpha=1, marker="*", edgecolor="black", linewidth=2)
plt.colorbar()
plt.xlabel("Day", fontsize=15)
plt.ylabel("Num", fontsize=15)
plt.title("Scatter Plot", fontsize=25)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
num=[2,3,4,2,4,5,6]


colors=[10,30,40,20,60,90,44]
sizes=[180,190,2800,155,380,2200,130]
plt.scatter(day, num, cmap="BrBg", s=sizes, alpha=1, marker="*", edgecolor="black", linewidth=2)
plt.colorbar()
plt.xlabel("Day", fontsize=15)
plt.ylabel("Num", fontsize=15)
plt.title("Scatter Plot", fontsize=25)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
num=[2,3,4,2,4,5,6]
num2=[3,8,6,12,22,4,16]


colors=[10,30,40,20,60,90,44]
sizes=[180,190,2800,155,380,2200,130]
sizes2=[120,220,800,255,300,2200,150]
plt.scatter(day, num, cmap="BrBg", s=sizes, alpha=1, marker="*", edgecolor="black", linewidth=2)
plt.scatter(day, num2, color="r", marker="+",s=sizes2)
t=plt.colorbar()
t.set_label("ColorBar")
plt.xlabel("Day", fontsize=15)
plt.ylabel("Num", fontsize=15)
plt.title("Scatter Plot", fontsize=25)
plt.show()
```
