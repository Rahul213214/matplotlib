# Stem Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,]
y=[2,3,5,4,9,7]

plt.stem(x, y, linefmt=":",markerfmt="r+", bottom=1)

plt.show()
```

```python
x=[1,2,3,4,5,6,]
y=[2,3,5,4,9,7]

plt.stem(x, y, linefmt=":",markerfmt="r+", bottom=9)

plt.show()
```

```python
x=[1,2,3,4,5,6,]
y=[2,3,5,4,9,7]

plt.stem(x, y, linefmt=":",markerfmt="r+", bottom=1, basefmt="g", label="Python", use_line_collection=False,
         orientation="horizontal")

plt.legend()

plt.show()
```
