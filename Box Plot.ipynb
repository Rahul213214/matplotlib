# Box Plot

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x)

plt.show()
```

```python

x=[10,20,30,40,50,60,70]

plt.boxplot(x, notch=True)

plt.show()
```

```python

x=[10,20,30,40,50,60,70]

plt.boxplot(x, vert=False)

plt.show()
```

```python

x=[10,20,30,40,50,60,70]

plt.boxplot(x, widths=0.8)

plt.show()
```

```python

x=[10,20,30,40,50,60,70]

plt.boxplot(x, labels=["Python"], patch_artist=True)

plt.show()
```

```python

x=[10,20,30,40,50,60,70]

plt.boxplot(x, showmeans=True)

plt.show()
```

```python

x=[10,20,30,40,50,60,70,150]

plt.boxplot(x, sym="r+", boxprops=dict(color="r"), capprops=dict(color="yellow"), whiskerprops=dict(color="g"))

plt.show()
```

```python

x=[10,20,30,40,50,60,70,150]

plt.boxplot(x, whis=3)

plt.show()
```

```python

x=[10,20,30,40,50,60,70]
x1=[10,20,40,80,60,70,55]

y=[x, x1]
plt.boxplot(y, labels=["Python", "c++"],sym="r+", boxprops=dict(color="r"), capprops=dict(color="yellow"), whiskerprops=dict(color="g"))
#plt.boxplot(x1, sym="r+", boxprops=dict(color="r"), capprops=dict(color="yellow"), whiskerprops=dict(color="g"))

plt.show()
```
