# Bar Plot

```python
import  matplotlib.pyplot as plt
x=['python', 'c','java','javascript']
y=[80,20,40,30]
plt.xlabel("language", fontsize=18)
plt.ylabel("No.", fontsize=18)
plt.title("Bar Plot", fontsize=25)
c=['y','b','g','black','pink']
plt.bar(x, y, width=0.3, color=c, align="edge", label="bar plot", edgecolor="r", linewidth=2, linestyle=':', alpha=1)
plt.legend()
plt.show
```

```python
x=['python', 'c','java','javascript']
y=[80,60,40,30]
z=[20,40,32,25]
plt.xlabel("language", fontsize=18)
plt.ylabel("No.", fontsize=18)
plt.title("Bar Plot", fontsize=25)
c=['y','b','g','black']
c2=['b','pink','red','orange']
plt.bar(x, y, width=0.3, color=c, align="edge", label="bar plot", edgecolor="r", linewidth=2, linestyle=':', alpha=1)
plt.bar(x, z, width=0.3, color=c2, align="edge", label="bar plot", edgecolor="r", linewidth=2, linestyle=':', alpha=1)
plt.legend()
plt.legend()
plt.show
```

```python
x=['python', 'c','java','javascript']
import numpy as np
y=[80,20,40,30]
z=[20,40,32,25]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language", fontsize=18)
plt.ylabel("No.", fontsize=18)
plt.title("Bar Plot", fontsize=25)

c=['yellow']
c2=['r']

plt.bar(p, y, width, color=c, label="popularity1")
plt.bar(p1, z, width, color=c2, label="popularity2")

plt.xticks(p+width/2, x, rotation=20)
plt.legend()
plt.show
```

```python
x=['python', 'c','java','javascript']
y=[80,20,40,30]
plt.xlabel("language", fontsize=18)
plt.ylabel("No.", fontsize=18)
plt.title("Bar Plot", fontsize=25)
c=['y','b','g','black','pink']
plt.barh(x, y, color=c, align="edge", label="bar plot", edgecolor="r", linewidth=2, linestyle=':', alpha=1)
plt.legend()
plt.show
```
